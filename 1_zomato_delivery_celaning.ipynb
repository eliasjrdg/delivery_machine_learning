# **Zomato Delivery Data Cleaning**

**Obtenção, Tratamento e Limpeza dos Dados:** Este notebook concentra-se em adquirir, tratar e limpar os dados, garantindo sua qualidade e preparação para as etapas seguintes.

```python
import pandas as pd
import os
```

# Importando os dados

```python
path = os.getcwd()
```

```python
df = pd.read_csv(os.path.join(path, 'data/Zomato Dataset.csv'))
df
```

```python
df.info()
```

# Data Cleansing

### Removendo registro com valores nulos

```python
df[df.isna().any(axis=1)]
```

```python
df_clean = df.dropna().copy()
df_clean.info()
```

### Tratando os registro de latitude e longitude

```python
df_clean['Restaurant_latitude'].value_counts()
```

* 3304 registros possuem valor de 'Restaurant_latitude' = 0. Vamos remover qualquer registro de localização com valor 0.

```python
df_clean = df_clean[(df_clean['Restaurant_latitude'] != 0) & (df_clean['Restaurant_longitude'] != 0) & (df_clean['Delivery_location_latitude'] != 0) & (df_clean['Delivery_location_longitude'] != 0)]
df_clean
```

```python
df_clean['Restaurant_latitude'].value_counts()
```

### Tratando a coluna 'Order_Date' para datetime

```python
df_clean['Order_Date'].value_counts()
```

```python
pd.options.mode.chained_assignment = None
df_clean['Order_Date'] = pd.to_datetime(df_clean['Order_Date'], format='%d-%m-%Y', errors='coerce')
df_clean.head()
```

```python
df_clean.info()
```

### Tratando as colunas de horas

```python
df_clean['Time_Orderd'].value_counts().sort_index()
```

```python
df_clean['Time_Order_picked'].value_counts().sort_index()
```

* Vários registros possuem valores diferentes do padrão de hora. Vamos definir um padrão 00:00 e remover os valores diferentes desse formato.

```python
time_pattern = r'^\d{2}:\d{2}$'
df_clean = df_clean[df_clean['Time_Orderd'].str.match(time_pattern)]
df_clean = df_clean[df_clean['Time_Order_picked'].str.match(time_pattern)]
df_clean.head(6)
```

```python
df_clean['Time_Order_picked'].value_counts().sort_index()
```
